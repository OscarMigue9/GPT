```python
'''Midlatitude Summer'''
import importlib
import final_forward_attenuation
importlib.reload(final_forward_attenuation)
from final_forward_attenuation import run_simulation, default_species, convert_atm, convert_vmr
```

Usando Vmrs en vez de presiones

```python
Ptotal = convert_atm(1.013e6, "gcms2") 
T = 294.20
ppVs = [1.876E-02, 3.300E-04, 3.017E-08, 3.200E-07, 1.500E-07, 1.700E-06, 0.781] #porque no usamos N2?
o2 = convert_vmr(2.090E+05)
vmrs = ppVs + [o2]
print("Temperatura:", T)
print("Presion total:", Ptotal)
print("Volumen mixing ratios:", vmrs)
```

```python
presiones = [Ptotal * vmr for vmr in vmrs]
```

```python
presiones
```

```python
'''Comprobando con la firma de H2O'''
sp = default_species()
for s in sp:
     if s.name == 'H2O': s.Pmol = vmrs[0] * Ptotal
     if s.name == 'CO2': s.Pmol = vmrs[1] * Ptotal
     if s.name == 'O3': s.Pmol = vmrs[2] * Ptotal
     if s.name == 'N2O': s.Pmol = vmrs[3] * Ptotal
     if s.name == 'CO': s.Pmol = vmrs[4] * Ptotal
     if s.name == 'CH4': s.Pmol = vmrs[5] * Ptotal
     if s.name == 'O2': s.Pmol = vmrs[7] * Ptotal

h2o = [s for s in sp if s.name == "H2O"]

import os
print(os.getcwd())

res = run_simulation(
    parfile='/home/jleon13/Documents/AFOSR project/Py4catsForward/Py4cats_Tinkering/680HIT87b.par',
    species=h2o,
    nu_min=666.67, nu_max=10000.0, dnu=0.01,
    tileW=20.0, guard=50.0,
    temp_K=T, L_m=1.0, pres=Ptotal,
    delta_um=0.020,
    save_csv=True, outdir='out', make_plots=True, att=False
)

lambda_um = res['lambda_centers']
Tprod = res['T_prod_samp']
Th2o  = res['T_each_samp'][0]
```
